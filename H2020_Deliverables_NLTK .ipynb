```python
# Selecting the projects which have the words textiles or clothing in the objectives
reports['larticle'] = reports['article'].str.lower()
terms = ['textile','textiles','clothing']
terms = '|'.join(terms)
r_textile = reports.loc[reports['larticle'].str.contains(terms, na=False),:]
r_textile.shape
```

```python
# tokenize the objectives of each project
# I used RegexpTokenize to eliminate punctuation, ...
r_textile.fillna("missing", inplace=True)
articles = r_textile[['projectAcronym','programme', 'topics','larticle']].values.tolist()


for i in range(0,len(articles)-1):
    row = articles[i][3]

    tokenizer = RegexpTokenizer(r'\w+\D+.\D+')      
    tokens = nltk.word_tokenize(row)
    stopWords = set(stopwords.words('english'))
    par_cleaned = [token for token in tokens if token not in stopWords] 

    NN=[]
    for element in par_cleaned:
        words = tokenizer.tokenize(element) 
        tagged = nltk.pos_tag(words)
        if tagged:
            if "NN" in tagged[0][1]:
                NN.append(tagged[0][0])
    articles[i].append(Counter(NN).most_common(25))
        
```

```python
# narrowing based on the fact that textiles is mentionned in the 25 most frequent words
relevant_frequent_words = ['textile', 'textiles','clothing','fibres','fibers','fashion']
relevant_projects =[]
for i in range(0,len(articles)-1):
    for j in range(0,24):
        #print(articles[i][4][j][1])
        if articles[i][4][j][0] in relevant_frequent_words:
            relevant_projects.append(articles[i])

relevant_projects_df = pd.DataFrame[relevant_projects]
```
